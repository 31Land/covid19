### <span style="color:darkred"> Define Features </span>

```python
INDs = Corona[['Age_up_to_17_Rate', 
               'Urban_Rate',
               'ICU_and_CCU'
               ]]

df = Corona[['Age_up_to_17_Rate', 
               'Urban_Rate',
               'ICU_and_CCU'
               ]]

# INDs = Corona[['Age_up_to_17_Rate', 
#                'Urban_Rate',
#                'ICU_and_CCU',
#                'HighVolumeArea'
#                ]]

# df = Corona[['Age_up_to_17_Rate', 
#                'Urban_Rate',
#                'ICU_and_CCU',
#                'HighVolumeArea'
#                ]]

dep = Corona.Diagnosed_Farvardin3.values
```

### <span style="color:darkred"> Sklearn + Metrics</span>

```python
# Sklearn: 
from sklearn.linear_model import LinearRegression
    
# fit_intercept to be as default TRUE to use .intercept for B0 coeeficent
# NOT including column of ONEs to get rid of having 0 as a first elemnt of B vectore
linreg = LinearRegression(fit_intercept=True,
                                   normalize=False,
                                   copy_X=True,
                                   n_jobs=1)  
model_SKlearn = linreg.fit(INDs, dep)
    
print(f"Coefficients(Beta Vectors):")
print(f"{model_SKlearn.coef_}")
print('---------------------------------------------------')
    
print(f"One unit increase in the first variable which is '{INDs.columns[0]}' and holding constant other independant features, corresponds to a {model_SKlearn.coef_[0]} unit increase in the estimate of DEPENDANT variable")
print('---------------------------------------------------')
    
print('\033[1m')
print(f"Shape of the coefficient is {model_SKlearn.coef_.shape}, so the number of features (p) are: {model_SKlearn.coef_.shape[0]}")
print('\033[0m')
print('---------------------------------------------------')
print('---------------------------------------------------')
    
print(f"Intercept:")
print(f"{model_SKlearn.intercept_}")
print('---------------------------------------------------')
print('---------------------------------------------------')
    
predictedS_SKlearn = model_SKlearn.predict(INDs)
print(f"Predicted values(y_hat)")
print(f"{predictedS_SKlearn}")
print('---------------------------------------------------')
print('\033[1m')
print(f"Shape of the Predicted value (which is dependant) of an independant variable(s) is {predictedS_SKlearn.shape}, so the number of observations (n) are: {predictedS_SKlearn.shape[0]}")
print('\033[0m')
print('---------------------------------------------------')
print(f"Error for the first predicted value against the true value: {predictedS_SKlearn[0]-dep[0]}")
print('---------------------------------------------------')
print('---------------------------------------------------')
    
RSquared = model_SKlearn.score(INDs, dep)
print(f"R-Squared by SKlearn:")
print(f"{RSquared}")
print('---------------------------------------------------')
print('---------------------------------------------------')
    
from sklearn.metrics import mean_squared_error, r2_score
import sklearn.metrics as metrics

MSE = metrics.mean_squared_error(dep, predictedS_SKlearn)
RMSE = metrics.mean_squared_error(dep, predictedS_SKlearn) ** 0.5
MedAE = metrics.median_absolute_error(dep, predictedS_SKlearn)
MeanAE = metrics.mean_absolute_error(dep, predictedS_SKlearn)
RSquared = metrics.r2_score(dep, predictedS_SKlearn)
    
p = INDs.shape[1] # Number of Features
n = INDs.shape[0] # Number of Observations
def r2_adj(dep, predictedS_SKlearn, p, n): 
    dep_mean = np.mean(dep)
    ssregadj = np.sum(np.square(dep - predictedS_SKlearn)) / (n-p-1)
    sstotadj = np.sum(np.square(dep - dep_mean)) / (n-1)
    return 1 - ssregadj / sstotadj
    
print(f"Mean Squared Error (MSE) by SKlearn:")
print(f"{MSE}")
print('---------------------------------------------------')
print('---------------------------------------------------')
print(f"ROOT Mean Squared Error (RMSE) by SKlearn:")
print(f"{RMSE}")
print('---------------------------------------------------')
print('---------------------------------------------------')
print(f"Median Absolute Error (MedAE) by SKlearn:")
print(f"{MedAE}")
print('---------------------------------------------------')
print('---------------------------------------------------')
print(f"Mean Absolute Error (MeanAE) by SKlearn:")
print(f"{MeanAE}")
print('---------------------------------------------------')
print('---------------------------------------------------')
print(f"R-Squared by SKlearn:")
print(f"{RSquared}")
print('---------------------------------------------------')
print('---------------------------------------------------')
print(f"R-Squared-Adjusted by SKlearn:")
print(f"{r2_adj(dep, predictedS_SKlearn, p, n)}")
print('---------------------------------------------------')
print('---------------------------------------------------')
        
print('\033[1m')
print(f"In the formula Y = BETA * X")
print('\033[0m')

print(f"Shape of the dependant vector (Y) should be 1 by n(number of observations), and here is:  {dep.shape}")
print(f"Shape of the BETA Coefficients (BETA) should be 1 by p(number of features, and here is:  {model_SKlearn.coef_.shape}")
print(f"Shape of the Independent Matrix (X) should be n(number of observations) by p(number of features), and here is:  {INDs.shape}")
```

### <span style="color:darkred"> Statsmodel </span>

```python
# statsmodel
import statsmodels.api as sm
        
# 'Statsmodel' doesn't authomatically add "intercept, so need to "Manually make a new column for the ...
# ... intercept in a design matrix (X), here is INDs
# Now, weather INDs is a dataframe or array, now is going to be a numpy array    
INDs = np.concatenate([INDs, np.ones((INDs.shape[0], 1))], axis=1)
    
# the dep variable comes before the Indipendant predictors    
model_SM = sm.OLS(dep, INDs)
    
model_SM = model_SM.fit()
predictedS_SM = model_SM.predict(INDs)
    
print('---------------------------------------------------')
print('---------------------------------------------------')
MSE_SM = mean_squared_error(dep, predictedS_SM)
print(f"Mean Squared Error (MSE) by Stats Model:")
print(f"{MSE_SM}")
print('---------------------------------------------------')
print('---------------------------------------------------')
    
print(model_SM.summary())
    
# Plotting the predicted values against the true values, using statsmodel
plt.figure(figsize=(10,8))
plt.scatter(predictedS_SM, dep, s=30, c='#f08807', marker='+', zorder=10)
plt.xlabel("Predicted Values - $\hat{y}$")
plt.ylabel("Actual Values - $y$")
plt.title("Plotting the predicted values against the true values, using statsmodel", fontdict=None, loc='center', pad=None)
plt.show();
```

### <span style="color:darkred"> Correlation HeatMap </span>

```python
def correlation_heat_map(df):
    corrs = df.corr()
    
    fig, ax = plt.subplots(figsize=(11,7))
    mask = np.zeros_like(corrs, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
        
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    ax = sns.heatmap(corrs, mask=mask, annot=True, cmap=cmap, vmin=-1, vmax=1)
        
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=10, rotation=30)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=10, rotation=0)
        
    plt.show()
        
df_with_target = df.copy()
df_with_target['dep'] = dep
correlation_heat_map(df_with_target)
```

### <span style="color:darkred"> Extra - K-Fold & Cross Validation </span>

```python
k = 5
```

```python
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
from sklearn import metrics
    
# iterating through folds 5-10
for folds in range(k,k+5):
    print ('------------------------------------\n')
    print(f"K: {folds}")
    model_KFold = LinearRegression()
    print('---------------------------------------------------')
    print('---------------------------------------------------')
        
    # Performing cross-validation
    scores_KFold = cross_val_score(model_KFold, INDs, dep, cv=folds)
    print(f"Cross-Validated Scores: {scores_KFold}")
    print('---------------------------------------------------')
    print('---------------------------------------------------')
    print(f"Mean CV R2: {np.mean(scores_KFold)}")
    print('---------------------------------------------------')
    print('---------------------------------------------------')
    print(f"Std CV R2: {np.std(scores_KFold)}")
    print('---------------------------------------------------')
    print('---------------------------------------------------')
                
    # Makeing cross-validated predictions
    predictionS_KFold = cross_val_predict(model_KFold, INDs, dep, cv=folds)
        
    RSquared_KFold = metrics.r2_score(dep, predictionS_KFold)
    print(f"Cross-Predicted R-Sqaured: {RSquared_KFold}")
    print('---------------------------------------------------')
    print('---------------------------------------------------')
        
# package which is useful for combinations and permutations
from itertools import combinations
    
# creating a list of all possible combinations of the Independent varaibles.
combs = []
for i in range(1, len(df.columns)+1):
    for c in combinations(df.columns.tolist(), i):
        combs.append(c)
            
# Optimizing R2
R2_results = {}
model_R2Optimize = LinearRegression()
KK = 250
for i,c in enumerate(combs):
    if (i % KK) == 0:
        print (i) # counting mechanism, showing the number of combinations iterred through
    subX = df[list(c)]
    R2_results[c] = np.mean(cross_val_score(model_R2Optimize, subX, dep))
# using every IV combination in the model
# getting the average of said results from the cross validation
# adding the average scores to the r2_results dictionary
    
R2_results_itemized = R2_results.items()
    
# converts dict to list of lists and sorts on the second item, in each lis, which is the R2 scores
R2_results_sorted = sorted(R2_results_itemized, key=lambda x: x[1], reverse=True)
    
# Converting List to Pandas Dataframe, sorting and showing the first 10 highest combination of features
R2_results_sorted = pd.DataFrame(R2_results_sorted, columns = ['Features' , 'Score'])
print(R2_results_sorted.sort_values('Score', ascending=False).head(10))
```
